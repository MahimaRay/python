Sales and Profit Analysis by State

```python
State_sales_profit = df.groupby('State', as_index=False)[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)

sns.barplot(x='Profit', y='State', data=State_sales_profit, color='red' ,  label='Profit' )
sns.barplot(x='Sales', y='State', data=State_sales_profit, color='yellow' ,  label='Sales' )
plt.xlabel('Profit')
plt.ylabel('State')
plt.title('Profit by State')
plt.show(10)
```

```python
profit_payment_mode = df.groupby('Payment Mode', as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Payment Mode', y='Profit', data=profit_payment_mode , color='orange')
plt.xlabel('Payment Mode')
plt.ylabel('Profit')
plt.title('Profit by Payment Mode')
plt.xticks(rotation=45)
plt.show()
```

 Sales and Profit Analysis by Sub-Category

a. Bar plot of total sales and profit by sub-category

```python
sub_category_sales_profit = df.groupby('Sub-Category', as_index=False)[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
plt.figure(figsize=(14, 7))
sns.barplot(x='Sub-Category', y='Sales', data=sub_category_sales_profit, color='blue', label='Sales')
sns.barplot(x='Sub-Category', y='Profit', data=sub_category_sales_profit, color='red', label='Profit', alpha=0.7)
plt.xlabel('Sub-Category')
plt.ylabel('Amount')
plt.title('Total Sales and Profit by Sub-Category')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

b. Line chart showing trends over time

```python
# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group by order date and calculate total sales and profit
date_sales_profit = df.groupby('Order Date')[['Sales', 'Profit']].sum()

# Plotting
plt.figure(figsize=(14, 7))
sns.lineplot(data=date_sales_profit)
plt.xlabel('Order Date')
plt.ylabel('Amount')
plt.title('Sales and Profit Trends Over Time')
plt.grid(True)
plt.show()
```

Sales and Profit Analysis by Region

```python
# Group by 'Region' and calculate total sales and profit
region_sales_profit = df.groupby('Region', as_index=False)[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=region_sales_profit, color='blue', label='Sales')
sns.barplot(x='Region', y='Profit', data=region_sales_profit, color='red', label='Profit', alpha=0.7)
plt.xlabel('Region')
plt.ylabel('Amount')
plt.title('Total Sales and Profit by Region')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

 Top Performing Products

```python
# Group by 'Product Name' and calculate total sales and profit
product_sales_profit = df.groupby('Product Name', as_index=False)[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False).head(10)

# Plotting
plt.figure(figsize=(14, 7))
sns.barplot(x='Sales', y='Product Name', data=product_sales_profit, color='blue', label='Sales')
sns.barplot(x='Profit', y='Product Name', data=product_sales_profit, color='red', label='Profit', alpha=0.7)
plt.xlabel('Amount')
plt.ylabel('Product Name')
plt.title('Top 10 Performing Products by Sales and Profit')
plt.legend()
plt.show()
```

Correlation Analysis

```python
# Select relevant numerical features
numerical_features = df[['Sales', 'Quantity','Profit' , 'Returns']]

# Calculate correlation matrix
correlation_matrix = numerical_features.corr()

# Plotting
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
